```python
import networkx as nx
from networkx.drawing.nx_agraph import write_dot, graphviz_layout
import matplotlib.pyplot as plt
```

```python
D = nx.scale_free_graph(10)
```

```python
filename = "10_0_0"
pos = graphviz_layout(D)
# plt.figure(figsize=(15,10))
nx.draw(D, pos=pos, with_labels=True, node_size=1000, node_color='c')
plt.tight_layout()
plt.savefig(filename+".png")
write_dot(D, filename+".dot")

plt.show()
```

```python
D.add_edge(0,9)
D.add_edge(2,8)
D.add_edge(1,4)
D.add_edge(0,6)
D.add_edge(1,7)
D.add_edge(1,5)
D.add_edge(7,3)
# D.remove_edge(1,0)
# D.add_edge(5,1)
# D.add_edge(5,9)
# D.add_edge(7,1)
```
