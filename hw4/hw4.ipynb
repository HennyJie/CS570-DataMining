```python
import networkx as nx
from networkx.drawing.nx_agraph import write_dot, graphviz_layout
import matplotlib.pyplot as plt
```

```python
D = nx.scale_free_graph(5)
```

```python
D.remove_edge(4,1)
# D.remove_edge(1,0)
D.add_edge(1,4)
```

```python
filename = "5_0_0"
pos = graphviz_layout(D)
# plt.figure(figsize=(15,10))
nx.draw(D, pos=pos, with_labels=True, node_size=1000, node_color='c')
plt.tight_layout()
plt.savefig(filename+".png")
write_dot(D, filename+".dot")

plt.show()
```
